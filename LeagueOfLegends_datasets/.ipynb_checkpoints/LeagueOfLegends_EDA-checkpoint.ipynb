League Of Legends EDA

We have data from 184k LoL matches downloaded from kaggle.
Our main research question is: do teams with classic role specialization win more often?

Classic role specialization means that teams try to pick balanced champion portfolio: 
by Class: Tank/Support, Fighter, Marksman, Assasin , Mage (mid)
or
by Role: TOP, MID, Support, Jungle, ADC

Other variations - non-classical approach. 
We are specifically interested (will compare against classics) when players pick non-traditional class-role combination
(using Tank as Jungle, or ADC as Jungle).

```python
# df_champs_short        # short champ

df = pd.read_csv(path_to_matches+'champs.csv')
df_champs_short = df


# df_team_composition    # team content - ss?
df = pd.read_csv(path_to_matches+'participants.csv')
df_team_composition = df


# df_team_milestones     # game milestones by team (killing baron)
df = pd.read_csv(path_to_matches+'teamstats.csv')
df_team_milestones = df



# df_champ_bans          # champ bans
df = pd.read_csv(path_to_matches+'teambans.csv')
df_champ_bans = df



# df_match          # idk
df = pd.read_csv(path_to_matches+'matches.csv')
df_matches = df
```

```python
# table addons (adding champs for example)
df_champ_bans = df_champ_bans.merge(df_champs_short, how='inner', left_on='championid', right_on='id')\
                            .drop('id', axis=1)

```

```python
ttl_matches = df_matches.gameid.nunique()
```

##### What are top 10 banned champs

```python
ttl_bans_by_champ = df_champ_bans.groupby('name', as_index=False)\
    .agg({'matchid':'nunique'})\
    .sort_values(by='matchid', ascending=False)
```

```python
champ_ban_order = df_champ_bans.groupby(['name','banturn'], as_index=False)\
    .agg({'matchid':'nunique'})\
    .sort_values(by='matchid', ascending=False)\
    .rename(columns={'matchid':'match_count'})
```

```python
# creating a pivot table of banned champs
a = champ_ban_order.pivot(index='name', columns='banturn', values='match_count' ).reset_index()


# mapping total bans to the right of the table
bans_order_by_champs = a.merge(ttl_bans_by_champ, how='inner', on='name')\
                        .rename(columns={'matchid':'ttl_bans'})
```

```python
# bans_order_by_champs['ban_rate'] = 
bans_order_by_champs.sort_values(by='ttl_bans', ascending=False).head(10)
```
