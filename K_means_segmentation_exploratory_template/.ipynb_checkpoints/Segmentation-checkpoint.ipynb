```python
## Imports
import pandas as pd
import numpy as np
import plotly.express as px

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#### Uploading data (and checking/correcting types


```python
df = pd.read_csv('online_retail_II.csv', encoding='windows-1251')
df.InvoiceDate = pd.to_datetime(df.InvoiceDate)
df['month'] = pd.to_datetime(df['invoicedate']).dt.to_period('M')
```

```python
# changing column names for more "comfortable" ones
new_cols = {}

for i in df.columns:
    new_cols[i] = i.lower().replace(' ','_')
df = df.rename(columns=new_cols)
```

```python
df.head()
```

```python
df.info()
```

#### Preparing data for RFM

```python
last_date = df.invoicedate.max()
```

```python
# interested in gmv only (also, the share of returns (negative values) in revenue is less than 1%)
df_prep = df[df.price > 0].groupby(['customer_id'], as_index=False)\
    .agg({'invoicedate':lambda x: (last_date - x.max()).days    # Recency
          , 'invoice':'nunique'                                      # Frequency - count of orders
          , 'price':'sum'})\
    .rename(columns={'invoicedate':'recency','price':'monetary','invoice':'frequency'})
```

```python
df_prep.info()
```

```python
df_prep.customer_id.nunique()
```

```python
quantiles = df_prep.quantile(q=[0.25,0.5, 0.75] )
quantiles
```

```python
rfmSegmentation = df_prep
```

```python
def RC_class(value, parameter_nm, quantile_table):
    if value <= quantile_table[parameter_nm][0.25]:
        return 1
    elif value <= quantile_table[parameter_nm][0.5]:
        return 2
    elif value <= quantile_table[parameter_nm][0.75]:
        return 3
    else:
        return 4

    
def FMClass(value, parameter_nm, quantile_table):
     if value <= quantile_table[parameter_nm][0.25]:
        return 4
     elif value <= quantile_table[parameter_nm][0.5]:
        return 3
     elif value <= quantile_table[parameter_nm][0.75]:
        return 2
     else:
        return 1
```

```python
rfmSegmentation['R_quartile'] = rfmSegmentation.recency.apply(RC_class, args=('recency',quantiles))
rfmSegmentation['F_quartile'] = rfmSegmentation.frequency.apply(FMClass, args=('frequency',quantiles))
rfmSegmentation['M_quartile'] = rfmSegmentation.monetary.apply(FMClass, args=('monetary',quantiles))

rfmSegmentation['RFM'] = rfmSegmentation['R_quartile'].map(str) +\
                         rfmSegmentation['F_quartile'].map(str) +\
                         rfmSegmentation['M_quartile'].map(str)
```

```python
path = 'rfm_result.csv'
rfmSegmentation.to_csv(path)
# link to tableu public dashboard: https://public.tableau.com/shared/298T5TR69?:display_count=n&:origin=viz_share_link
```
